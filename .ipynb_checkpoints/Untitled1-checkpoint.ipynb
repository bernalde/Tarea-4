```python
filelist=glob.glob('./Brahe-f/*.dat')

if(os.path.isfile('results.txt')):
    os.remove('results.txt')
    os.remove('matrix.dat')
    os.remove('vector.dat')
temp=open('matrix.dat','w')
data=np.loadtxt(filelist[0])
b=data[:,0]
c=b**2/2
a=np.ones(size(b))
temp.write('38 3\n')
for j in range(size(b)):
    temp.write('%f %f %f\n'%(a[j],b[j],c[j]))
temp.close
count=0;
for i in filelist:
    data=np.loadtxt(i)
    d=data[:,1]
    temp2=open('vector.dat','w')
    temp2.write('38 1\n')
    for j in range(size(b)):
        temp2.write('%f \n'%(d[j]))
    temp2.close
    x=!./t4.out matrix.dat vector.dat
    y0=float(x[0])
    v0=float(x[1])
    a=float(x[2])
    r2=float(x[3])
    fileout=open('results.txt','a')
    if(count>0):
        name=filelist[count-1].strip('./Brahe-f/experiment_theta_')
        name=name.rstrip('.dat')
        name=name.rsplit('_phi_')
        theta=float(name[0])
        phi=float(name[1])
        fileout.write('%f %f %f %f %f %f\n'%(theta,phi,y0,v0,a,r2))
    count+=1
fileout.close
```

Punto 1. Se generó un archivo de text 'results.txt' que contiene 6 columnas. En la primera el ángulo polar ($\theta$), en la segunda el ángulo azimutal ($\phi$), en la tercera el valor inicial de altura ($y_0$), en la cuarta la velocidad inicial ($v_0$), en la quinta la aceleración de ese punto con eje de referencia positivo en dirección radial ($a$) y en la quinta el coeficiente de correlación  entre los valores experimentales y aquellos predichos por el ajuste (r²).

```python
results=np.loadtxt('results.txt')
theta=results[:,0]
phi=results[:,1]
y0=results[:,2]
v0=results[:,3]
a=results[:,4]
r2=results[:,5]
scatter(theta,a)
ax = axes()
ax.set_xlabel("$theta$",fontsize=20)
ax.set_ylabel("$a$",fontsize=20)
ax.set_title("$\mathrm{Gravity\ vs\ polar\ angle}$", fontsize=20)
filename = 'figure1'
```

```python
plot(r2)
```

Puede notarse que existen puntos cuya linealización fue mala (r² bajo) por lo que se descartan dichos valores.

```python
j=0;
a_clean=np.zeros(0)
theta_clean=np.zeros(0)
for r in r2:
    if(r>0.9):
        a_clean=np.append(a_clean,a[j])
        theta_clean=np.append(theta_clean,theta[j])
    j+=1
#En este paso se aplica un filtro sobre los datos que tengan un coeficiente de correlación r² menor a 0.9
#El r² se calcula con la comparación de los datos experimentales y lo determinado por experimentos
scatter(theta_clean,a_clean)
ax = axes()
ax.set_xlabel("$theta$",fontsize=20)
ax.set_ylabel("$a$",fontsize=20)
ax.set_title("$\mathrm{Gravity\ vs\ polar\ angle}$", fontsize=20)
filename = 'figure2'
```

Punto 2. Relizando un filtrado a partir del valor del coeficiente de correlación de los datos experimentales y aquellos predichos por el modelo, se obtiene una gráfica de aceleración de la gravedad en función del ángulo polar.

```python
suma=0
count=1
for g in a_clean:
    suma+=g
    count+=1
g_prom=suma/count
F=1-a_clean/g_prom
scatter(theta_clean,F)
ax = axes()
ax.set_xlabel("$theta$",fontsize=20)
ax.set_ylabel("$F$",fontsize=20)
ax.set_title("$\mathrm{Gravity\ variation\ vs\ polar\ angle}$", fontsize=20)
filename = 'figure3'
```
